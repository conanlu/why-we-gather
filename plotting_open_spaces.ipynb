## Plotting open spaces as polygons

```python
import pandas as pd
import requests
import json
import numpy as np
import time as tm
import geopy.distance
from matplotlib import pyplot as plt
import geopandas
import geodatasets
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
import plotly.io as pio
pio.renderers.default = "notebook"
import plotly.graph_objs as go
```

```python
open_space_source = 'data/raw/Open_Space.geojson'
f = open(open_space_source)
data = json.load(f)
```

```python
# fixing lat/long
for i in range(len(data['features'])):
    list1 = data['features'][i]['geometry']['coordinates'][0][0]
    for j in range(len(list1)):
        temp = list1[j][0]
        list1[j][0] = list1[j][1]
        list1[j][1] = temp
```

```python
mapbox_t=open(".mapbox_token").read().rstrip()
data_initialization = [go.Scattermapbox(
                    lat=[42.3295],
                    lon=[-71.0902],
)]
#set the layout to plot
layout = go.Layout(autosize=True,
                   mapbox = dict(center= dict(lat=42.3295, 
                                 lon=-71.0902),        
                                 accesstoken=mapbox_t,
                                 zoom=11,
                                 style='light',
                               ),
                   
                    width=1000,
                    height=800,
                    title = 'Boston')

fig = go.Figure(data=data_initialization, layout=layout)
```

```python
mylayers =[]

layer= dict(sourcetype = 'geojson',
             source = data,
             below='',  
             type = 'fill',
             opacity = 0.2,
             color = "green")

mylayers.append(layer)
```

```python
fig.layout.update(mapbox_layers = mylayers);
```
