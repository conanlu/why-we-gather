```python
# TODO: probably abstract into a function or functions so it can be done repeatedly
import numpy as np
import pandas as pd
import datetime
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
trips = []
for i in range(1, 25):
    trips.append(pd.read_pickle(f'data/bluebikes/trips_combined_boston/{i}.pkl'))
combined_trips = pd.concat(trips)
combined_trips
```

```python
STATION_NAME = 'Mass Ave T Station'
```

```python
station = combined_trips.loc[(combined_trips['start_station_name'] == STATION_NAME) | (combined_trips['end_station_name'] == STATION_NAME)]
station
```

```python
HOURS = [datetime.time(i) for i in range(24)]
HOURS_FORMATTED = [datetime.time(i).strftime('%-I %p') for i in range(24)]
```

```python
def plot_daily_activity(df, station_name):
    started = df.loc[df['start_station_name'] == station_name]
    ended = df.loc[df['end_station_name'] == station_name]

    b1 = plt.hist(started['started_at'].dt.time, alpha=0.5, bins=24, label='Departures')
    b2 = plt.hist(ended['ended_at'].dt.time, alpha=0.5, bins=24, label='Arrivals')
    plt.xticks(HOURS[::4], HOURS_FORMATTED[::4])
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Number of Arrivals or Departures')
    plt.title(f'Hourly Bluebike Arrivals and Departures, {station_name}')
    plt.show()

    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(np.linspace(0, 2*np.pi, 25), list(b1[0]) + [b1[0][0]], marker='.', label='Departures')
    ax.plot(np.linspace(0, 2*np.pi, 25), list(b2[0]) + [b2[0][0]], marker='.', label='Arrivals')
    ax.set_theta_direction(-1)
    ax.set_theta_zero_location("N")
    ax.set_xticks(np.linspace(0, 23/12*np.pi, 24), HOURS_FORMATTED)
    plt.legend()
    plt.title(f'Hourly Bluebike Arrivals and Departures, {station_name}')
    plt.show()
```

```python
plot_daily_activity(station, STATION_NAME)
```

```python
station_year_trunc = station.loc[station['ended_at'].dt.date < datetime.date(2024, 1, 1)]
```

```python
b3 = plt.hist(station_year_trunc['ended_at'].dt.dayofyear, bins=365//5)
xax = plt.gca().xaxis
xax.set_major_locator(mdates.MonthLocator())
xax.set_major_formatter(mdates.DateFormatter('%b'))
plt.title(f'Average (Daily?) Rides During the Year, {STATION_NAME}')
plt.show()
```
