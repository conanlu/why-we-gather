```python
import pandas as pd
import plotly.express as px
px.set_mapbox_access_token(open(".mapbox_token").read())
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
street_trees = pd.read_csv('data/locations/raw/street_trees.csv')
street_trees
```

```python
fig = px.scatter_mapbox(street_trees, lat='Y', lon='X', opacity=0.6,
                        center={'lat': 42.312, "lon": -71.0902}, zoom=11, width=1080, height=720)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
